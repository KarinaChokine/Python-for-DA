````python
import numpy as np
import pandas as pd

np.random.seed(0)

products = ['apple', 'banana', 'cherry', 'dates', 'elderberry']
categories = ['fruit', 'fruit', 'fruit', 'fruit', 'fruit']

N_ROWS = 10**6

df = pd.DataFrame({
    'product_name': np.random.choice(products, N_ROWS),
    'product_category': np.random.choice(categories, N_ROWS),
    'product_price': np.random.uniform(1, 10, N_ROWS),
    'transaction_date': pd.date_range(start='1/1/2000', periods=N_ROWS, freq='h')
})

memory_before = df.memory_usage(deep=True).sum() / (1024**2)

df['product_name'] = df['product_name'].astype('category')
df['product_category'] = df['product_category'].astype('category')

memory_after = df.memory_usage(deep=True).sum() / (1024**2)
ratio = memory_before / memory_after

print(f"Пам'ять до: {memory_before:.2f} MB")
print(f"Пам'ять після: {memory_after:.2f} MB")
print(f"Новий df використовує пам'яті у {ratio:.2f} разів менше")
````

- імпортувала бібліотеки в алфавітному порядку

- використала метод .astype('category') для ефективного перетворення текстових стовпців на категорійний тип

- додала підрахунок пам'яті за допомогою memory_usage(deep=True).sum(), щоб врахувати реальну пам'ять, яку займають рядки (strings)

- перевела байти в мегабайти (поділивши на 10242) для кращої читабельності

- додала розрахунок співвідношення (ratio), щоб побачити, у скільки разів зменшився обсяг

````python
def calculate_variance(numbers):
    # 1. Знаходимо загальну суму всіх чисел у списку
    sum_list = 0
    for n in numbers:
        sum_list += n
        
    # 2. Обчислюємо середнє арифметичне (математичне сподівання)
    mean = sum_list / len(numbers)
    
    # 3. Знаходимо суму квадратів усіх чисел
    sum_squares = 0
    for n in numbers:
        sum_squares += n**2
        
    # 4. Обчислюємо середнє значення квадратів
    mean_squares = sum_squares / len(numbers)
    
    # 5. Дисперсія дорівнює різниці між середнім квадратом і квадратом середнього
    return mean_squares - mean**2


# Задаємо тестовий список оцінок/значень
input_list = [75, 80, 85, 90, 95]

# Викликаємо функцію та виводимо результат 
print(calculate_variance(input_list))
````
