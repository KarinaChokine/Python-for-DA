1. Запис параметрів

```python
sample_mean = data.mean()
usual_mean = 20
usual_std = np.std(data, ddof=1)
sample_size = 100
alpha = 0.05
```

2. Формування гіпотез

H₀: μ = 20 (середній чек не змінився)


H₁: μ > 20 (середній чек значущо підвищився)


Тип тесту: правосторонній (one-tailed)

3. Обчислення стандатрної помилки SE

```python
standard_error = usual_std / np.sqrt(sample_size)
print(f'SE = {standard_error}')
```

4. Розрахунок z-статистики

```python
z_statistic = (sample_mean - usual_mean) / standard_error
print(f'z-статистика = {z_statistic}')
print()
if z_statistic > 1.2:
    print('Це досить велике відхилення')
elif z_statistic > 0.9:
    print('Це помірне відхилення')
else :
    print('Це невелике відхилення')
```

5. Пошук z-критичного

```python
z_critical = stats.norm.ppf(1 - alpha)
print(f'z-критичне = {z_critical}')
```

6. Обчислення p-value

```python
p_value = 1 - stats.norm.cdf(z_statistic)
print(f'p-value = {p_value}')
```

7. Висновки

```python
print(f"z-статистика = {z_statistic:.3f}")
print(f"Критичне значення = {z_critical:.3f}")
print(f"Порівняння: {z_statistic:.3f} {'>' if z_statistic > z_critical else '≤'} {z_critical:.3f}")

if z_statistic > z_critical:
    print("Висновок: z-статистика ПЕРЕВИЩУЄ критичне значення")
    decision1 = "ВІДХИЛЯЄМО H₀"
else:
    print("Висновок: z-статистика НЕ перевищує критичне значення")
    decision1 = "НЕ ВІДХИЛЯЄМО H₀"

print(f"Рішення: {decision1}")
```

```python
z_stat_sm, p_val_sm = stests.ztest(data,          
                                    value = usual_mean, 
                                    alternative = 'larger')   
print(f'z-статистика = {z_stat_sm}')
print(f'p-значення = {p_val_sm}')
print()
if p_val_sm < alpha :
    print('Відхиляємо H₀')
else :
    print('Не відхиляємо H₀')
```
