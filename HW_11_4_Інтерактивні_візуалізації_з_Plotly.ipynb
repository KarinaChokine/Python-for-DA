## Завдання 3: Порівняння Plotly Express vs Graph Objects

**Завдання:**
Створіть лінійний графік помісячної динаміки оренди (кількість оренд) велосипедів двома способами - з Plotly Express та з Graph Objects.

**Дайте відповіді на питання.**
1. Як ви розумієте основну різницю між цими двома підходами?
2. Коли краще використовувати Plotly Express?
3. Коли потрібен Graph Objects?


```python
df_monthly = df.resample('ME', on='datetime')['count'].sum().reset_index()
```

```python
px.line(
    df_monthly, 
    x='datetime', 
    y='count',
    title='Помісячна динаміка оренди (Plotly Express)'
)
```

```python
fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x = df_monthly['datetime'],
        y = df_monthly['count'],
        mode = 'markers + lines'
    )
)
        
```

1. Експрес він більше для швидкого аналізу, багато роботи він робить сам 'під капотом'. Наприклад він автоматично підписує вісі, коли го такого не робить.  Go більше для тонких налаштувань.
2. Краще використовувати коли потрібно швидко подивитися на дані, і не так важливо як це все виглядає.
3. Для тонких налаштувань, для складних дашбордів.

## Завдання 4 (Опціональне): Дашборд з make_subplots (Graph Objects)

**Завдання:**
Створіть дашборд з 4 різними графіками в одній фігурі:
- Bar chart - середні значення загальної кількості оренд велосипедів за сезонами(кварталами)
- Pie chart - відсоткове співвідношення погодних умов в даних
- Line chart - середнє значення загальної кількості оренд велосипедів за годинами протягом доби
- Scatter plot - кореляція температури vs вологість

Додайте заголовок на дашборд.

**Дайте відповідь на питання**
- На ваш погляд, яка перевага об'єднання графіків в один дашборд?

```python
fig = make_subplots(
    rows = 2, cols = 2,
    specs=[[{"type": "bar"}, {"type": "pie"}],
           [{"type": "scatter"}, {"type": "scatter"}]],
    subplot_titles = ('Оренда за сезонами', 'Погодні умови', 'Середня оренда по годинам', 'Кореляція температури vs вологість')
)

season_data = df.groupby('season')['count'].mean().reset_index()
fig.add_trace(
    go.Bar(
    x = season_data['season'],
    y = season_data['count'],
    name = 'Оренда',
    hovertemplate = 'Сезон: %{x}<br>Середня оренда: %{y}<extra></extra>'
),
row = 1, col = 1
)

df_weather = df['weather'].value_counts().reset_index()
fig.add_trace(
    go.Pie(labels=df_weather['weather'], values=df_weather['count'], 
        name='Погода', 
      hovertemplate = 'Погода: %{label}<br>Відсоток: %{value}<extra></extra>'), 
    row=1, col=2
)

df_hourly = df.groupby('hour')['count'].mean().reset_index()
fig.add_trace(
    go.Scatter(
        x = df_hourly['hour'],
        y = df_hourly['count'],
        name = 'Оренда',
        hovertemplate = 'Година: %{x}<br>Середня оренда: %{y}<extra></extra>'
    ),
    row = 2, col = 1
)

fig.add_trace(
    go.Scatter(
        x = df['temp'],
        y = df['humidity'],
        mode = 'markers',
        hovertemplate = 'Температура: %{x}<br>Вологість: %{y}<extra></extra>'
       
    ),
    row = 2, col = 2
)

fig.update_layout(
    height = 800,
    showlegend = False,
    title_text = 'Дашборд з make_subplots'
)
fig.show()
```

Можна побачити цілу картину, і в сусідньому графіку можна знайти відповідь на питання яке утворилося з іншого дашборду.
